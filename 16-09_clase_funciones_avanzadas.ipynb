### Ejercicio 1
Crear una función recursiva llamada cuenta_adelante(n) que imprima los numeros del 1 al n

```python
def cuenta_adelante(n):
  if n== 0:
    return
  else: 
    cuenta_adelante(n - 1)
    print(n)
```

```python
cuenta_adelante(3)
```

### Ejercicio 2
Crea una función recusrisva llamada potencia(base,exponente) que calcule base exponente


```python
def potencia(base, exponente):
  if exponente == 0: #Caso base
    return 1
  else:
    return base * potencia(base, exponente-1)
```

```python
potencia(2,4)
```

```python
def num_cifras (numero):
  if numero < 10:
    return 1 #caso recursivo
  else:
    return 1 + num_cifras(numero // 10)
  
```

```python
print(num_cifras(1452))
```

```python
def invertir_cadena(cadena):
  if len(cadena) == 0:
    return "" 
  else:
    return cadena[-1] + invertir_cadena (cadena[:-1])
    
  
```

```python
print(invertir_cadena('python'))
```
